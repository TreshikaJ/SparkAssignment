```python
#Read a text file and count the number of words in the file using RDD operations.
txt_file=spark.sparkContext.textFile("/content/sampleTemp.txt")
words=txt_file.flatMap(lambda d:d.split(" ")).collect()
words
count=txt_file.flatMap(lambda d:d.split(" ")).count()
count
```

```python
#Write a program to find the word frequency in a given file.
w_freq=txt_file.flatMap(lambda d:d.split(" ")).map(lambda d:(d,1)).reduceByKey(lambda a,b:a+b).collect()
w_freq
```

```python
#Write a program to convert all words in a file to uppercase.
uppercase=txt_file.map(lambda d:d.upper())
uppercase.collect()
```

```python
#Write a program to convert all words in a file to lowercase.
lowercase=txt_file.map(lambda d:d.lower())
lowercase.collect()
```

```python
#Write  a  program  to  capitalize  first  letter  of  each  words  in  file
def isAlphabet(d):
    return d.isalpha()

capital=txt_file.map(lambda d:d.capitalize()).collect()
capital
```

```python
#Mapthe  Registration  numbers to corresponding  branch. 58000 series  BDA,57000  series AIML,
#38000 series VLSI, 39000 series ES, and 47000 series CDC.
#Given registration number,generate a key-value pair of Registration Number and Corresponding Branch.
def reg_to_branch(reg):
  if 58000<=reg<59000:
    return "BDA"
  if 57000<=reg<58000:
    return "AIML"
  if 38000<=reg<39000:
    return "VLSI"
  if 39000<=reg<40000:
    return "ES"
  if 47000<=reg<48000:
    return "CDC"
reg=[39678,58900,57111,47899,39455]
reg_no=spark.sparkContext.parallelize(reg)
map=reg_no.map(lambda d:(d,reg_to_branch(d)))
map.collect()
```

```python
#Text file contain numbers. Numbers are separated by one white space. There is no order to store the numbers.
#One line may contain one or more numbers. Find the maximum, minimum, sum and mean of numbers.
numbers=spark.sparkContext.textFile("/content/numbers.txt")
max=numbers.flatMap(lambda d:d.split(' ')).filter(lambda d:len(d)>0).map(lambda d:int(d)).max()
min=numbers.flatMap(lambda d:d.split(' ')).filter(lambda d:len(d)>0).map(lambda d:int(d)).min()
sum=numbers.flatMap(lambda d:d.split(' ')).filter(lambda d:len(d)>0).map(lambda d:int(d)).sum()
mean=numbers.flatMap(lambda d:d.split(' ')).filter(lambda d:len(d)>0).map(lambda d:int(d)).mean()
print(max,min,sum,mean)
```

```python
#.A text file(citizen.txt)contains data about citizens of country.
#Fields(information in file) are Name, dob, Phone, email and state name.
#Another file contains mapping of state names to state code like  Karnataka  is  codes  as  KA,
#TamilNadu  as  TN,  Kerala  KL  etc.
#Compress  the citizen.txt file  by changing full state name to state code.
stateRDD = spark.sparkContext.textFile('/content/codes.txt')
stateKey=stateRDD.map(lambda d:(d.split(' ')[0],d.split(' ')[1]))
stateKey.collect()
codes_dict={}
for val in stateKey.collect():
  codes_dict[val[0]]=val[1]
codes_dict



def changeToCode(d,codes_dict):
  new_data=' '
  split_data=d.split(' ')
  split_data[4]=codes_dict.get(split_data[4])
  new_data=new_data.join(split_data)
  return new_data


citizens=spark.sparkContext.textFile('/content/citizen1.txt')
mapCitizens=citizens.map(lambda d:changeToCode(d,codes_dict))
mapCitizens.collect()
```
